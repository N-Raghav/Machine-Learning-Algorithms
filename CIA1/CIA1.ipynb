```python
from sklearn.preprocessing import LabelEncoder
```

```python
def object_to_int(dataframe_series):
    if dataframe_series.dtype=='object':
        dataframe_series = LabelEncoder().fit_transform(dataframe_series)
    return dataframe_series
```

```python
data
```

```python
data = data.apply(lambda x: object_to_int(x))
data.head()
```

```python
data=data.drop(['customerID'],axis=1)
```

```python
data.apply(lambda x: x.max() - x.min())
```

```python
cols=['tenure','MultipleLines','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','Contract','PaymentMethod','MonthlyCharges','TotalCharges']
```

```python
#since we need to use algorithms like Decision Tree,Naive Bayes which have logarithm 
#in the formula, we need to scale the data using Standard Scaler
```

```python
data['Churn'] = data['Churn'].astype(int)
```

```python
x=data.iloc[:,0].values
y=data.iloc[:,1].values
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df = pd.DataFrame(scaler.fit_transform(data), columns=data.columns)
```

```python
df=df.drop(['Churn'],axis=1)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
# K-Nearest Neighbors (KNN)
knn_classifier = KNeighborsClassifier()
X_train_reshaped = x_train.reshape(-1, 1)
X_test_reshaped = x_test.reshape(-1, 1)
knn_classifier.fit(X_train_reshaped, y_train)
knn_predictions = knn_classifier.predict(X_test_reshaped)
knn_accuracy = accuracy_score(y_test, knn_predictions)
print(f'KNN Accuracy: {knn_accuracy:.4f}')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=2)
```

```python
# Logistic Regression
logistic_classifier = LogisticRegression()
X_train_reshaped = x_train.reshape(-1, 1)
X_test_reshaped = x_test.reshape(-1, 1)
logistic_classifier.fit(X_train_reshaped, y_train)
logistic_predictions = logistic_classifier.predict(X_test_reshaped)
logistic_accuracy = accuracy_score(y_test, logistic_predictions)
print(f'Logistic Regression Accuracy: {logistic_accuracy:.4f}')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=3)
```

```python
# Naive Bayes
nb_classifier = GaussianNB()
X_train_reshaped = x_train.reshape(-1, 1)
X_test_reshaped = x_test.reshape(-1, 1)
nb_classifier.fit(X_train_reshaped, y_train)
nb_predictions = nb_classifier.predict(X_test_reshaped)
nb_accuracy = accuracy_score(y_test, nb_predictions)
print(f'Naive Bayes Accuracy: {nb_accuracy:.4f}')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=4)
```

```python
# Decision Trees
tree_classifier = DecisionTreeClassifier()
X_train_reshaped = x_train.reshape(-1, 1)
X_test_reshaped = x_test.reshape(-1, 1)
tree_classifier.fit(X_train_reshaped, y_train)
tree_predictions = tree_classifier.predict(X_test_reshaped)
tree_accuracy = accuracy_score(y_test, tree_predictions)
print(f'Decision Trees Accuracy: {tree_accuracy:.4f}')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=5)
```

```python
# Support Vector Machines (SVM)
svm_classifier = SVC()
X_train_reshaped = x_train.reshape(-1, 1)
X_test_reshaped = x_test.reshape(-1, 1)
svm_classifier.fit(X_train_reshaped, y_train)
svm_predictions = svm_classifier.predict(X_test_reshaped)
svm_accuracy = accuracy_score(y_test, svm_predictions)
print(f'SVM Accuracy: {svm_accuracy:.4f}')
```
