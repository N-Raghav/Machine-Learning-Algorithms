```python
#preprocessing given in the problem statement
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None)

dataset = dataset.fillna(0)

transactions = []

for i in range(0, 7501):

    transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])
```

```python
# Function to generate candidate itemsets of size k
def generate_candidates(prev_candidates, k):
    candidates = []
    for i in range(len(prev_candidates)):
        for j in range(i+1, len(prev_candidates)):
            # Take first k-2 items from both itemsets
            itemset1 = prev_candidates[i][:-1]
            itemset2 = prev_candidates[j][:-1]

            # If the first k-2 items are equal, join the sets
            if itemset1 == itemset2:
                new_candidate = tuple(list(prev_candidates[i]) + [prev_candidates[j][-1]])  # Convert list to tuple
                # Check if the candidate has frequent subsets
                if has_infrequent_subset(new_candidate, prev_candidates, k-1):
                    continue
                candidates.append(new_candidate)
    return candidates

# Function to check if all subsets of a candidate are frequent
def has_infrequent_subset(candidate, prev_candidates, k):
    subsets = get_subsets(candidate, k)
    for subset in subsets:
        if subset not in prev_candidates:
            return True
    return False

# Function to generate subsets of size k from a candidate
def get_subsets(candidate, k):
    subsets = []
    subset_generator(candidate, 0, [], subsets, k)
    return subsets

# Recursive function to generate subsets
def subset_generator(candidate, index, current_subset, subsets, k):
    if len(current_subset) == k:
        subsets.append(current_subset)
        return
    if index >= len(candidate):
        return
    subset_generator(candidate, index+1, current_subset + [candidate[index]], subsets, k)
    subset_generator(candidate, index+1, current_subset, subsets, k)

# Function to count support for each candidate
def count_support(transactions, candidates):
    support = defaultdict(int)
    for transaction in transactions:
        for candidate in candidates:
            if set(candidate).issubset(set(transaction)):
                support[candidate] += 1
    return support

# Function to filter candidates with minimum support
def filter_candidates(support, min_support):
    frequent_candidates = []
    for candidate, count in support.items():
        if count >= min_support:
            frequent_candidates.append(candidate)
    return frequent_candidates

# Function to generate frequent itemsets using Apriori algorithm
def apriori(transactions, min_support):
    frequent_itemsets = []
    k = 1
    # Generate frequent itemsets of size 1
    candidates = [tuple([item]) for item in set([item for transaction in transactions for item in transaction])]
    while candidates:
        # Count support for candidates
        support = count_support(transactions, candidates)
        # Filter candidates with minimum support
        frequent_candidates = filter_candidates(support, min_support)
        # Add frequent itemsets of size k to the result
        frequent_itemsets.extend(frequent_candidates)
        # Generate candidates of size k+1
        candidates = generate_candidates(frequent_candidates, k+1)
        k += 1
    return frequent_itemsets

# Set minimum support threshold
min_support = 100

# Run Apriori algorithm
frequent_itemsets = apriori(transactions, min_support)

# Print frequent itemsets
for itemset in frequent_itemsets:
    print(itemset)
```
